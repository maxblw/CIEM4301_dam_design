```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
H = 575 - 495
B = 100
Q_10000 = 9603 #m3/s
Q10 = 2259

dz = np.linspace(0.5*H, H, 10000)
Cd = 0.611 + (0.08 * ((H - dz) / dz))
Q = (2/3) * Cd * B * np.sqrt(2 * 9.81 * (H-dz)**3)
hc = (Q_10000**2 / (9.81 * B**2))**(1/3)


fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].plot(dz, Cd)
axes[0].set_title('$C_d$ vs. $\\Delta z$')
axes[0].set_xlabel('$\\Delta z$ (m)')
axes[0].set_ylabel('$C_d$')
axes[0].grid()

axes[1].plot(dz, Q)
axes[1].set_title('Discharge $Q$ vs. $\\Delta z$')
axes[1].set_xlabel('$\\Delta z$ (m)')
axes[1].set_ylabel('Discharge $Q$ (m$^3$/s)')
axes[1].set_yscale('log')
axes[1].grid()


dz_10000 = dz[np.argmin(np.abs(Q - Q_10000))]
print(f"Hd = {H} m")
print(f"hc = {hc:.2f} m")
print(f"Discharge at 10,000 m³/s occurs at Δz = {dz[np.argmin(np.abs(Q - Q_10000))]:.2f} m")

```
